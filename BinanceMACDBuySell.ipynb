```python
#main code which will run continously by calling neccessary function repeatedly

client = connection()

buy = 0

#list to hold prices values for buy and sell
buyList = []
sellList = []

#code will run continously untill we interrupt this manually
while(1):
  if buy == 0:
    time.sleep(17)
    macd = get_MACD()
    if macd['valueMACD'] < 0:
      #macd difference values taken  by trying and experiencing result
      if macd['valueMACD'] < macd['valueMACDSignal'] and (abs(macd['valueMACDSignal']) - abs(macd['valueMACD'])) <= 0.03:
        price = get_Real_Price('ETHUSDT',client)
        buy = 1
        buyList.append(price)
    else:
      if macd['valueMACD'] > macd['valueMACDSignal'] and (abs(macd['valueMACD']) -abs(macd['valueMACDSignal'])) <= 0.03:
        price = get_Real_Price('ETHUSDT',client)
        buy = 1
        buyList.append(price)
  elif buy == 1:
    time.sleep(20)
    macd = get_MACD()
    if macd['valueMACD'] < macd['valueMACDSignal']:
      price = get_Real_Price('ETHUSDT',client)
      buy = 0
      sellList.append(price)
  print('current price : ', get_Real_Price('ETHUSDT',client))
  print('buy : ',buyList)
  print('sell : ',sellList)

  
```
